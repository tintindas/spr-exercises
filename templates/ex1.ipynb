# Part 1: Matrix Inversion
Define a matrix and a vector and compute the result of a matrix-vector multiplication.
Invert the matrix.

```python
a = np.array([[1, 2],
[3, 4]])

print(f"Matrix A: {a}")

b = np.array([[5], [6]])

print(f"Vector B: {b}")


c = a @ b

print(f"Matrix-vector multiplication: {c}")

d = np.linalg.inv(a)

print(f"Inverse of Matrix: {d}")
```
