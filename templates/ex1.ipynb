# Part 1: Matrix Inversion
Define a matrix and a vector and compute the result of a matrix-vector multiplication.
Invert the matrix.

```python
a = np.array([[1, 2],
[3, 4]])
b = np.array([5, 6])

c = np.matmul(a, b)

print(f"C: {c}")

d = np.linalg.inv(a)

print(f"Result: {d}")
```
